```python
import seaborn as sns
```

```python
sns.load_dataset("glue")
```

```python
glue = sns.load_dataset("glue").pivot(index="Model",columns="Task",values ="Score")
glue
```

```python
sns.heatmap(glue)
```

```python
# xticklabels = False: sáº½ ko hiá»ƒn thá»‹ label cho trá»¥c x
sns.heatmap(glue, xticklabels=False,yticklabels=False)
```

```python
# Use annot to represent the cell values with text (thÃªm giÃ¡ trá»‹ vÃ o trong cÃ¡c Ã´ biá»ƒu Ä‘á»“):
sns.heatmap(glue, annot=True)
```

```python
# Control the annotations (chÃº thÃ­ch) with a formatting string (thÃªm pháº§n tháº­p phÃ¢n):
sns.heatmap(glue, annot=True, fmt=".1f")
```

```python
# hiá»ƒn thÃ¬ pháº§n chÃº thÃ­ch giÃ¡ trá»‹ rank theo cá»™t, vd dÃ²ng BERT tÆ°Æ¡ng á»©ng vs cá»™t CoLA thÃ¬ cÃ³ rank 1, cá»™t MNLI thÃ¬ rank 4,...)
sns.heatmap(glue, annot=glue.rank(axis="columns"))
```

```python
# Add lines between cells:
sns.heatmap(glue, annot=True, linewidth=.5)
```
